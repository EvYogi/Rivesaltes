# Block Matching

Illustration of how block matching is used in this library

```python
from smartgrid.image_processing.block_matching import BlockMatching
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
%matplotlib inline
```

```python
bm = BlockMatching()
```

## Import two consequent infrared images 

```python
img_1 = np.array(Image.open('./acceptance_tests/data/201804032345.jpg').convert('L'))
img_2 = np.array(Image.open('./acceptance_tests/data/201804040000.jpg').convert('L'))
```

```python
plt.imshow(img_1)
```

```python
Vx, Vy, X, Y = bm.block_match(img_1, img_2)
```

## Visualisation of raw calculation

```python
fig, ax = plt.subplots()
ax.quiver(X, Y, Vx, Vy)
plt.show()
```

```python
Vx, Vy = bm.smoothing_vectors(Vx, Vy)
```

## Visualisation of smoothed vectors

```python
fig, ax = plt.subplots()
ax.quiver(X, Y, Vx, Vy)
plt.show()
```

## Computation of interpolated vectors

```python
Vx, Vy, X, Y = bm.interpolate_vectors(Vx, Vy, X, Y)
```
